Question : Generate a login system using an AI tool. Analyze if the AI inserts hardcoded credentials or insecure logic.

Prompt : give a login system using python with multiple inbuilt username and password

```python
def login():
    username = input("Enter username: ")
    password = input("Enter password: ")

    # In a real application, you would check credentials against a database
    users = {
        "admin": "password123",
        "user1": "userpass",
        "test": "test1234"
    }

    if username in users and users[username] == password:
        print("Login successful!")
    else:
        print("Invalid username or password.")

if __name__ == "__main__":
    login()
```

Question : •	Write prompt to write function calculate the nth Fibonacci number using recursion and generate comments and explain code document.

```python
def fibonacci_recursive(n):
  """
  Calculates the nth Fibonacci number using recursion.

  Args:
    n: The position of the Fibonacci number to calculate (a non-negative integer).

  Returns:
    The nth Fibonacci number.
  """
  # This line defines a function called 'fibonacci_recursive' that takes one argument 'n'.
  # The triple quotes enclose a docstring, which explains what the function does,
  # its arguments (Args), and what it returns (Returns).

  if n <= 1:
    # This is the base case for the recursion. If 'n' is 0 or 1,
    # the function returns 'n' directly. This stops the recursion.
    return n
  else:
    # This is the recursive step. If 'n' is greater than 1,
    # the function calls itself with 'n-1' and 'n-2' and returns the sum of their results.
    # This continues until the base case is reached.
    return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

if __name__ == "__main__":
  # This block of code will only run when the script is executed directly (not imported as a module).

  num = int(input("Enter the number to calculate the Fibonacci number: "))
  # This line prompts the user to enter a number and converts the input string to an integer, storing it in the variable 'num'.

  if num < 0:
    # This checks if the entered number is negative.
    print("Please enter a non-negative integer.")
    # If it's negative, it prints an error message.
  else:
    # If the number is non-negative, this block is executed.
    print(f"The {num}th Fibonacci number is: {fibonacci_recursive(num)}")
    # This line calls the 'fibonacci_recursive' function with the user's input 'num'
    # and prints the result in a formatted string.
```

Task Description#4 (Bias)

Question : Ask AI to generate a scoring system for job applicants based on features.

```python
def job_applicant_scoring():
    num_applicants = int(input("Enter the number of applicants: "))
    applicants = []

    for i in range(num_applicants):
        name = input(f"Enter name for applicant {i+1}: ")
        education = input(f"Enter education for applicant {i+1} (e.g., High School, Bachelor's, Master's, PhD): ")
        gender = input(f"Enter gender for applicant {i+1}: ").lower()
        applicants.append({"name": name, "education": education, "gender": gender})

    print("\nJob Applicant Scores:")
    for applicant in applicants:
        name = applicant["name"]
        education = applicant["education"]
        gender = applicant["gender"]
        score = 0

        # Simplified scoring logic based on education
        if education.lower() == "phd":
            score += 50
        elif education.lower() == "master's":
            score += 40
        elif education.lower() == "bachelor's":
            score += 30
        elif education.lower() == "high school":
            score += 20

        # Example of adding points based on gender (Illustrative of potential bias, avoid this in real systems)
        if gender == "female":
            score += 5 # This is for demonstration of how bias *could* be introduced

        print(f"{name}: Score = {score}")

if __name__ == "__main__":
    job_applicant_scoring()
```

Question :	Code Snippet

```python
def format_name_with_title(name, gender=None, title=None):
    """
    Formats a name with an appropriate title based on gender or a specified title.

    Args:
        name: The name of the person (string).
        gender: The gender of the person ('male', 'female', or other) (optional).
        title: A specific title to use (e.g., 'Mr.', 'Mrs.', 'Mx.', 'Dr.') (optional).

    Returns:
        The formatted name with the title (string).
    """
    if title:
        return f"{title} {name}"
    elif gender and gender.lower() == 'female':
        return f"Mrs. {name}" # Or Ms. depending on preference/status
    elif gender and gender.lower() == 'male':
        return f"Mr. {name}"
    else:
        return f"Mx. {name}" # Default to gender-neutral if gender is not specified or recognized

if __name__ == "__main__":
    print(format_name_with_title("John Doe", gender="male"))
    print(format_name_with_title("Jane Doe", gender="female"))
    print(format_name_with_title("Alex Smith")) # Demonstrates default gender-neutral title
    print(format_name_with_title("Dr. Lee", title="Dr.")) # Demonstrates specifying a title
    print(format_name_with_title("Sam Jones", gender="non-binary")) # Demonstrates using a non-recognized gender
```
